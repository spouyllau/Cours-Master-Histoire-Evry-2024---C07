## Boucles `while` 
Une alternative à l'instruction for couramment utilisée en informatique est la boucle while. Avec ce type de boucle, une série d'instructions est exécutée tant qu'une condition est vraie. Par exemple :

````python
i = 1
while i <= 4:
    print(i)
    i = i + 1
````

````python
i = 1
while i <= 10:
    print(i)
    i = i + 2
````

````python
i = 0
while i > 10:
    print("Le Python c'est cool !")
````

Faites bien attention aux tests et à l'incrémentation que vous utilisez, car une erreur mène souvent à des « boucles infinies » qui ne s'arrêtent jamais. 

````python
i = 0
while i < 10:
    reponse = input("Entrez un entier supérieur à 10 : ")
    i = int(reponse)
````

## Les fonctions 

En programmation, les fonctions sont très utiles pour réaliser plusieurs fois la même opération au sein d'un programme. Elles rendent également le code plus lisible et plus clair en le fractionnant en blocs logiques.

Pour l'instant, une fonction est à vos yeux une sorte de « boîte noire » :

- À laquelle vous passez aucune, une ou plusieurs variable(s) entre parenthèses. Ces variables sont appelées arguments. Il peut s'agir de n'importe quel type d'objet Python.
- Qui effectue une action.
- Qui renvoie un objet Python ou rien du tout.



````python
len([0, 1, 2])
````

Voici ce qu'il se passe :

- vous appelez len() en lui passant une liste en argument (ici la liste [0, 1, 2]) ;
- La fonction calcule la longueur de cette liste ;
- Elle vous renvoie un entier égal à cette longueur.

Pour définir une fonction, Python utilise le mot-clé `def`. Si on souhaite que la fonction renvoie quelque chose, il faut utiliser le mot-clé `return`. Par exemple :

````python
def carre(x):
    return x*2
print(carre(2))
````

````python
def hello():
    print("Hello World")
hello()
````

### Fonctions et arguments de fonctions

````python
def fois(x, y):
    return x*y
fois(2, 3)
````

# Gestions des données CSV avec Pandas (_Dataframe_) et analyse avec matplotlib

**Pandas** est une bibliothèque Python open source pour la manipulation et l'analyse des données. Pandas se compose de structures de données et de fonctions directement utilisable en quelques lignes de python permettant d'effectuer des opérations sur les données.

**Matplotlib** est une bibliothèque de traçage de graphiques de bas niveau en python qui sert d'utilitaire de visualisation. Matplotlib, qui a été créée par John D. Hunter, est open source et peut être utilisée librement.

## Dataframe sous Pandas
Un DataFrame est un ensemble de valeurs sous la forme de _series Pandas_ indexées par une valeur.

````python
import pandas as pd 
import numpy as np

# A structured array
my_array = np.ones(3, dtype=([('foo', int), ('bar', float)]))
# Print the structured array
print(my_array['foo'])

# A record array
my_array2 = my_array.view(np.recarray)
# Print the record array
print(my_array2.foo)
````

````python
data = np.array([['','Col1','Col2'],
                ['Row1',1,2],
                ['Row2',3,4]])
                
print(pd.DataFrame(data=data[1:,1:],
                  index=data[1:,0],
                  columns=data[0,1:]))
````

Gestion du _Dataframe_ : 

````python
df = pd.DataFrame({"A":[1,4,7], "B":[2,5,8], "C":[3,6,9]})
print(df)
````

````python
# Sélectionner la ligne 0 avec iloc
print(df.iloc[0])

# Sélectionner la colonne A avec loc
print(df.loc[:,'A'])
````

Ajouter un index pour gérer mieux ses données :

````python
print(df)
df.set_index('C')
````
